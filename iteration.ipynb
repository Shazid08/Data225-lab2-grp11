```python
import pymongo
import json

# Connect to MongoDB
client = pymongo.MongoClient("mongodb+srv://shazidshaik:Shazid0000@lab2.pdvonc8.mongodb.net/?retryWrites=true&w=majority&appName=Lab2")
db = client["Lab2"]
collection = db["Mdata2"]

# Function to safely convert stringified JSON to nested objects with original keys
def convert_to_object(data):
    try:
        # Check if data is empty or None
        if not data:
            return None
        
        # If data is already a string, parse it directly
        if isinstance(data, str):
            return json.loads(data)
        # If data is a list, convert it to a JSON string and then parse it
        elif isinstance(data, list):
            return json.loads(json.dumps(data))
        else:
            print("Invalid data type. Expected string or list.")
            return None
    except json.JSONDecodeError as e:
        print("Error parsing JSON data:", e)
        return None

# Update documents in the collection
for document in collection.find():
    update_data = {
        "$set": {
            "genres": convert_to_object(document["genres"]),
            "production_companies": convert_to_object(document["production_companies"]),
            "production_countries": convert_to_object(document["production_countries"]),
            "cast": convert_to_object(document["cast"]),
            "crew": convert_to_object(document["crew"]),
            "keywords": convert_to_object(document["keywords"])
        }
    }
    collection.update_one({"_id": document["_id"]}, update_data)

print("Conversion completed successfully.")
```
